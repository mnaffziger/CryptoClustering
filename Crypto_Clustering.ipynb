```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
display(df_market_data.head(10))
df_market_data.info()
df_market_data.columns
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data)

# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(market_data_scaled, columns= ['price_change_percentage_24h', 
                                                                   'price_change_percentage_7d',
                                                                   'price_change_percentage_14d', 
                                                                   'price_change_percentage_30d',
                                                                   'price_change_percentage_60d', 
                                                                   'price_change_percentage_200d',
                                                                   'price_change_percentage_1y']
                                                                   )


# Copy the crypto names from the original data
df_market_data_scaled['coinid'] = df_market_data.index

# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index('coinid')

# Display sample data
display(df_market_data_scaled.head())

# View scaled sample data
df_market_data_scaled.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=2)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k,
              "inertia" : inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plot_elbow_orig = df_elbow.hvplot.line(
    x = "k",
    y = "inertia",
    title = "Elbow Curve of Original Fitted Data",
    xticks = k
)

plot_elbow_orig
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** At k = 4 the difference in inertia starts to level off.  I would start at k = 4 and evaluate the predicted clusters.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model_k4 = KMeans(n_clusters= 4, random_state= 1)

# Fit the K-Means model using the scaled data
model_k4.fit(df_market_data_scaled)

# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = model_k4.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
k_4[:20]
```

```python
# Create a copy of the DataFrame
df_market_data_pred_k4 = df_market_data_scaled.copy()

# Add a new column to the DataFrame with the predicted clusters
df_market_data_pred_k4['pred k= 4'] = k_4

# Display sample data
df_market_data_pred_k4.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

plot_orig_k_4 = df_market_data_pred_k4.hvplot.scatter(
    x = "price_change_percentage_24h",
    y = "price_change_percentage_7d",
    by= "pred k= 4",
    title= "Original Data Cluster Predictions, k = 4",
    hover_cols = 'coinid' # eventhough 'coinid' is dba the datafram index, since it is defined as 'coinid' I can just call it directly instead of 'index'
)

plot_orig_k_4
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)

# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_data_pca = pca.fit_transform(df_market_data_scaled)
# View the first five rows of the DataFrame. 
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** It explains the percentage of the data that is represented by the PCA model.  When the dimentions of this dataset were reduced, the model accounted for ~90% of the features from the original, scaled dataset.

```python
# Create a new DataFrame with the PCA data.
df_market_data_pca = pd.DataFrame(market_data_pca, columns=['PCA_1', 'PCA_2', 'PCA_3'])

# # Copy the crypto names from the original data
df_market_data_pca['coinid'] = df_market_data.index

# # Set the coinid column as index
df_market_data_pca = df_market_data_pca.set_index('coinid')

# Display sample data
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca = list(range(1,11))

# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k_pca:
    k_model = KMeans(n_clusters=i, random_state=2)
    k_model.fit(df_market_data_pca)
    inertia_pca.append(k_model.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {
    'k_pca' : k_pca,
    'inertia_pca' : inertia_pca
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)
# df_elbow_pca.head(10)

# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plot_elbow_pca_3 = df_elbow_pca.hvplot.line(
    x = 'k_pca',
    y = 'inertia_pca',
    title = 'Elbow Curve of PCA Data, Components = 3',
    xticks = k_pca
)

plot_elbow_pca_3
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** From the PCA elbow plot, `k= 4` still looks like the best value


* **Question:** Does it differ from the best `k` value found using the original data?

  * **Answer:** No

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_pca = KMeans(n_clusters=4)

# Fit the K-Means model using the PCA data
model_pca.fit(df_market_data_pca)

# Predict the clusters to group the cryptocurrencies using the PCA data
market_data_pca_clusters = model_pca.predict(df_market_data_pca)

# Print the resulting array of cluster values.
print(market_data_pca_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
df_market_data_pca_pred = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_market_data_pca_pred['pred_clusters'] = market_data_pca_clusters

# Display sample data
df_market_data_pca_pred.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

plot_pca_3_k_4_PCA12 = df_market_data_pca_pred.hvplot.scatter(
    x = 'PCA_1',
    y = 'PCA_2',
    by = 'pred_clusters',
    title = "PCA Fitted Model of Original Data. Compenents = 3, k = 4",
    hover_cols = 'coinid'
    )
plot_pca_3_k_4_PCA13 = df_market_data_pca_pred.hvplot.scatter(
    x = 'PCA_1',
    y = 'PCA_3',
    by = 'pred_clusters',
    title = "PCA Fitted Model of Original Data. Compenents = 3, k = 4",
    hover_cols = 'coinid'
    )
plot_pca_3_k_4_PCA23 = df_market_data_pca_pred.hvplot.scatter(
    x = 'PCA_2',
    y = 'PCA_3',
    by = 'pred_clusters',
    title = "PCA Fitted Model of Original Data. Compenents = 3, k = 4",
    hover_cols = 'coinid'
    )
(plot_pca_3_k_4_PCA12 + plot_pca_3_k_4_PCA13 + plot_pca_3_k_4_PCA23).cols(2)
```

  * **Answer:**  With this dataset, the PCA model dropped ~10% of the feature data that did not fit the model.  This resulted in a tighter clustering of the data to visualize the clustering better.  In both plots, coins ethlend and celsius-degree-token are both single members of different groups.  From the original scaled data plot, only coin ethlend is visualized easily as a separate cluster.  Coin celsius-degree-token is burried within another cluster and would be missed if it was not color coded. The PCA model also appears to have provided a cleaner, more distinct boundry between clusters 0 and 1 (blue and red respectfully in the PCA cart immediatly above) than the original data's cluster plot.

```python
#-----
# Additional analysis of which feature is strongly related to  / corresponds with each PCA component
# Determin which of the features correspond to each PCA component and visualize the weights assigned from the model
market_data_pca_components = pca.components_
df_market_data_pca_components = pd.DataFrame(market_data_pca_components, columns=['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d','price_change_percentage_60d', 'price_change_percentage_200d','price_change_percentage_1y'])
model_components = ['PCA_1', 'PCA_2', 'PCA_3']
df_market_data_pca_components['components'] = model_components
df_market_data_pca_components = df_market_data_pca_components.set_index('components')
display(df_market_data_pca_components.head())

# Convert the loadings DataFrame into a long format for easier plotting
df_weights = df_market_data_pca_components.reset_index().melt(id_vars='components')

# Rename the columns for clarity
df_weights.columns = ['pca_component', 'feature', 'weight']
df_weights['weight'] = df_weights['weight'].abs()

width = 800
height = 600
# Create the hvplot line plot
plot_pca_weights_bar = df_weights.hvplot.bar(
    x = 'feature',y='weight',
    by = 'pca_component', stacked = True,
    xlabel='PCA Component',ylabel='Absolute Value of Feature Weight',
    title= 'Feature Weight on Each PCA Component',
    rot = 45,
    height = height,
    width = width
)
plot_pca_weights_bar
```

#### PCA Component and Feature Relationships

All three compenents weighted the daily price change almost equally, with a higher absolute value of weight assigned to PCA_1 (0.417) and PCA_2 (0.358)

* **PCA_1:** Represents the variances related to more of the long term prices changes (>200 days).  This component could be seen as an annual price performance

* **PCA_2:** Represents the variances related to mid term price changes between 14 - 60 days.  This component could be seen as a quarterly price preformance

* **PCA_3:** Represents the variances related to shorter term price changes between 7 days.  This component could be seen as a weekly price preformance
